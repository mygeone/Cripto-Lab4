# **Clustering**

El objetivo de clustering consiste en formar grupos cerrados y homogéneos a partir de un conjunto de elementos que tienen diferentes características o propiedades, pero que comparten ciertas similitudes.

Existen una gran variedad de algoritmos de *clustering* con distintas aplicaciones y enfocados en distintos segmentos. Sin embargo, **KMeans** es el algoritmo de clustering mas utilizado actualmente.
Es un algoritmo de clasificación no supervisada que agrupa en k grupos basándose en sus características. El agrupamiento se realiza minimizando la suma de distancias entre cada objeto y el centroide de su grupo o cluster.

Podemos resumir el funcinoamiento de KMeans en:


1.   Inicializar
2.   Asignación objetos a los centroides
3.   Actualización posición centroides

El algoritmo descrito anteriormente se repite de forma recursiva hasta que los centroides no se muevan

# **Usos del clustering**

Uno de los principales usos del clustering, es la segmentación de clientes. El algoritmo provee una agrupación de datos en base a los clientes, que permite caracterizar a los clientes en base a su comportamiento. Esto lo hace especialmente util en analisis de mercado y clientes.

Al momento de hacer llamado de KMeans, se sabe por catedra que funciona sobre una cantidad definida de clusters y estos puede uno definirlos y asi mismo los datos quedan agrupados en funci[on de la distancia Euclidiana siendo esta última la que genera la densidad de categorización de los elementos según etiquetas y/o otros identificadores.

Para efectos de clustering, se elimina una variable que indica si la persona ha muerto durante el estudio. Un valor de 1 indica que murió en el periodo de estudio y 0 indica que no.

```python
kmeans = KMeans(n_clusters=2)
dataDropped = df.drop(['DEATH_EVENT'],axis=1)
```

Sin los datos a predecir, se realiza clustering

```python
kmeans.fit(dataDropped)
dataClusters = pd.concat([dataDropped, pd.DataFrame(kmeans.labels_, columns=['IsDeath'])], axis=1)
```

 Una vez realizado el clustering, se procede a realizar una predicción respecto al dato eliminado. Para esto utilizaremos regresión lineal

```python
############ prediction
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.rcParams['figure.figsize'] = (16, 15)
plt.style.use('ggplot')
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train = np.array(dataDropped) # Features
Y_train = dataClusters['IsDeath'].values # Target
```

Con los datos seleccionados que utilizamos anteriormente en el proceso de clustering, realizamos la regresión lineal

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)

y_pred = regr.predict(X_train)
```

Con la predicción realizada, podemos analizar los distintos coeficientes asociados a la predicción

```python
# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('Coefficients: \n', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regr.intercept_)
# Error Cuadrado Medio
print("Mean squared error: %.2f" % mean_squared_error(Y_train, y_pred))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Variance score: %.2f' % r2_score(Y_train, y_pred))

'''
 preds = list()

for i in range(len(dataDropped)):
  a = (dataDropped.iloc[i])
  y_Dosmil = regr.predict([a])
  preds.append(int(y_Dosmil))

r2_score(dataClusters[['IsDeath']],preds) 
'''
```

##Ejercicio 2: Regresion Lineal

```python
# Imports necesarios
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#cargamos los datos de entrada
data = pd.read_csv("./sample_data/data.csv")
#veamos cuantas dimensiones y registros contiene
data.shape
```

```python
#son 161 registros con 8 columnas. Veamos los primeros registros
data.head()
```

```python
# Ahora veamos algunas estadísticas de nuestros datos
data.describe()
```

```python
data.drop(['time', 'DEATH_EVENT'],1).hist()
plt.show()
```

```python
# Vamos a RECORTAR los datos en la zona donde se concentran más los puntos
# esto es en el eje X: entre 0 y 3.500
# y en el eje Y: entre 0 y 80.000
filtered_data = data[(data['age'] <= 100) & (data['creatinine_phosphokinase'] <= 7861)]

colores=['green','blue']
tamanios=[30,60]

f1 = filtered_data['age'].values
f2 = filtered_data['creatinine_phosphokinase'].values

# Vamos a pintar en colores los puntos por debajo y por encima de la media de Cantidad de Palabras
asignar=[]
for index, row in filtered_data.iterrows():
    if(row['age']>65):
        asignar.append(colores[0])
    else:
        asignar.append(colores[1])
    
plt.scatter(f1, f2, c=asignar, s=tamanios[0])
plt.show()
```

```python
# Asignamos nuestra variable de entrada X para entrenamiento y las etiquetas Y.
dataX =filtered_data[["age"]]
X_train = np.array(dataX)
y_train = filtered_data['creatinine_phosphokinase'].values

# Creamos el objeto de Regresión Linear
regr = linear_model.LinearRegression()

# Entrenamos nuestro modelo
regr.fit(X_train, y_train)

# Hacemos las predicciones que en definitiva una línea (en este caso, al ser 2D)
y_pred = regr.predict(X_train)

# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('Coefficients: \n', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regr.intercept_)
# Error Cuadrado Medio
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Variance score: %.2f' % r2_score(y_train, y_pred))
```

```python
#Vamos a comprobar:
# Quiero predecir cuántos "Shares" voy a obtener por un artículo con 2.000 palabras,
# según nuestro modelo, hacemos:
y_Dosmil = regr.predict([[50]])
print(int(y_Dosmil))
```

```python
#Vamos a intentar mejorar el Modelo, con una dimensión más: 
# Para poder graficar en 3D, haremos una variable nueva que será la suma de los enlaces, comentarios e imágenes
suma = (filtered_data["creatinine_phosphokinase"] + filtered_data['high_blood_pressure'].fillna(0) + filtered_data['platelets'])

dataX2 =  pd.DataFrame()
dataX2["age"] = filtered_data["age"]
dataX2["suma"] = suma
XY_train = np.array(dataX2)
z_train = filtered_data['time'].values
```

```python
# Creamos un nuevo objeto de Regresión Lineal
regr2 = linear_model.LinearRegression()

# Entrenamos el modelo, esta vez, con 2 dimensiones
# obtendremos 2 coeficientes, para graficar un plano
regr2.fit(XY_train, z_train)

# Hacemos la predicción con la que tendremos puntos sobre el plano hallado
z_pred = regr2.predict(XY_train)

# Los coeficientes
print('Coefficients: \n', regr2.coef_)
# Error cuadrático medio
print("Mean squared error: %.2f" % mean_squared_error(z_train, z_pred))
# Evaluamos el puntaje de varianza (siendo 1.0 el mejor posible)
print('Variance score: %.2f' % r2_score(z_train, z_pred))
```

```python
fig = plt.figure()
ax = Axes3D(fig)

# Creamos una malla, sobre la cual graficaremos el plano
xx, yy = np.meshgrid(np.linspace(0, 100, num=10), np.linspace(0, 60, num=10))

# calculamos los valores del plano para los puntos x e y
nuevoX = (regr2.coef_[0] * xx)
nuevoY = (regr2.coef_[1] * yy) 

# calculamos los correspondientes valores para z. Debemos sumar el punto de intercepción
z = (nuevoX + nuevoY + regr2.intercept_)

# Graficamos el plano
ax.plot_surface(xx, yy, z, alpha=0.2, cmap='hot')

# Graficamos en azul los puntos en 3D
ax.scatter(XY_train[:, 0], XY_train[:, 1], z_train, c='blue',s=30)

# Graficamos en rojo, los puntos que 
ax.scatter(XY_train[:, 0], XY_train[:, 1], z_pred, c='red',s=40)

# con esto situamos la "camara" con la que visualizamos
ax.view_init(elev=30., azim=65)
        
ax.set_xlabel('creatinine_phosphokinase')
ax.set_ylabel('high_blood_pressure')
ax.set_zlabel('platelets')
ax.set_title('Regresión Lineal con Múltiples Variables')
```
